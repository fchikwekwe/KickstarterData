```python
# Checking all the columns to see if they have null or NaN values 
df["ID"].isna().sum() # 0
df["name"].isna().sum() # 4, We will need to examine these null values
df["category"].isna().sum() # 0
df["main_category"].isna().sum() # 0
df["currency"].isna().sum() # 0
df["deadline"].isna().sum() # 0
df["goal"].isna().sum() # 0
df["launched"].isna().sum() # 0
df["pledged"].isna().sum() # 0
df["state"].isna().sum() # 0
df["backers"].isna().sum() # 0
df["country"].isna().sum() # 0
df["usd pledged"].isna().sum() # 3797, We need to check on these values as well
df["usd_pledged_real"].isna().sum() # 0 
df["usd_goal_real"].isna().sum() # 0 
```

```python
# Lets check the 4 rows that have NaN in the "name" column 
no_name = df[df["name"].isna()]
no_name
```

```python
# It looks like these rows still have valuable data, so lets just fill the NaNs with harmless values
df["name"] = df["name"].fillna("No Name")
```

```python
# No more NaNs! 
df["name"].isna().sum()
```

```python
# Now lets check on the usd pledged NaNs 

df["ID"].isna().sum() # 0
df["name"].isna().sum() # 4, 🤔
df["category"].isna().sum() # 0
df["main_category"].isna().sum() # 0
df["currency"].isna().sum() # 0
df["deadline"].isna().sum() # 0
df["goal"].isna().sum() # 0
df["launched"].isna().sum() # 0
df["pledged"].isna().sum() # 0
df["state"].isna().sum() # 0
df["backers"].isna().sum() # 0
df["country"].isna().sum() # 0
df["usd pledged"].isna().sum() # 3797
df["usd_pledged_real"].isna().sum() # 0 
df["usd_goal_real"].isna().sum() # 0 
```

```python
state_undefined = df[df.state == "undefined"]
state_undefined
```

```python
# checking to see what values we have in other columns when state == undefined 

state_undefined["country"].unique() # returns array(['N,0"'], dtype=object)
state_undefined["usd pledged"].unique() # returns array([nan])
state_undefined["backers"].unique() # returns array([0])

# all other columns has a normal assortment of values
```

## It seems that state, country, usd pledged and backers all have null or zero values in these ~3500 to 4000 rows 
## These rows still appear to have other values that are normal, so we do not want to just drop the row. 

## Now I know that: 
### Null values in 'state' column are 'undefined'
### Null values in 'country' column are 'N,0"'

## I can fill the NaN values in usd pledged

```python
# I'm going to fill the NaNs in usd pledged with the vaue in usd_pledged_real since both are simply 
# values translated to USD. Even though the exchange rates might be slightly different, and the values do
# deviate in some rows, they are generally pretty close to one another. 

df["usd pledged"] = df["usd pledged"].fillna(df["usd_pledged_real"]).astype(int)
```

```python
# Much better! 
df["usd pledged"].isna().sum()
```
